```python
import pandas as pd

df = pd.read_csv("academia_redfit.csv")

df
```

```python
df.dropna()
```

```python
df.dropna(subset=["sexo", "idade", "frequencia_semanal_treino", "tipo_atividade", "tempo_medio_exercicio", "minutos_totais_semana", "preco_plano", "possui_nutricionista", "primeiro_PGC", "ultimo_PGC", "data_matricula", "estado"])
```

```python
df.fillna(0)
```

```python
df.drop_duplicates()
```

```python
colunas = ["sexo", "idade", "frequencia_semanal_treino", "tipo_atividade", 
           "tempo_medio_exercicio", "minutos_totais_semana", "preco_plano", 
           "possui_nutricionista", "primeiro_PGC", "ultimo_PGC", 
           "data_matricula", "estado"]

df[colunas] = df[colunas].apply(lambda x: x.str.strip() if x.dtype == 'object' else x)
```

```python
coluna_numericas = ["minutos_totais_semana", "frequencia_semanal_treino"]
df[coluna_numericas] = df[coluna_numericas].apply(lambda x: x.astype(int) if x.dtype == 'float64' else x)
```

```python
df
```

```python
df["sexo"] = df["sexo"].str.strip().str.lower().map({
    "masculino": "masculino",
    "m": "masculino",
    "homem": "masculino",
    "h": "masculino",
    "feminino": "feminino", 
    "f": "feminino",
    "mulher": "feminino",
    "outro": "outro",
    "o": "outro",
    "outros": "outro"
})
```

```python
df['data_matricula'] = pd.to_datetime(df['data_matricula'], format='%Y-%m-%d')
```

```python
df
```

```python
df.to_csv('dados_tratados.csv', index=False)
```
