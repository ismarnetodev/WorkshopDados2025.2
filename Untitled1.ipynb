```python
import pandas as pd

df = pd.read_csv("/content/sample_data/dados_nao_tratados.csv")

print(df.info())
print(df.head())
```

```python

df["nota"] = pd.to_numeric(df["nota"], errors="coerce")
df["nota"] = df["nota"].fillna(df["nota"].mean())

df["idade"] = pd.to_numeric(df["idade"], errors="coerce")
df["idade"] = df["idade"].fillna(df["idade"].median())
```

```python
df["ativo"] = df["ativo"].str.strip().str.lower()
df["ativo"] = df["ativo"].map({"sim": True, "não": False, "true": True, "e": False})
```

```python
df["data_inscricao"] = pd.to_datetime(df["data_inscricao"], errors="coerce")
```

```python
df = df.drop_duplicates()
```

```python
df.rename(columns={
    "id": "id",
    "nome": "nome",
    "idade": "idade",
    "data_inscricao": "data_inscricao",
    "nota": "nota",
    "ativo": "ativo"
}, inplace=True)
```

```python
df.to_csv("dados_tratados.csv", index=False)
```
